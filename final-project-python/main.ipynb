```python
from database import get_engine, get_main_table, get_intraday_archive_table
from history_data import initial_history_all
import datetime


def setup(create_history: bool = False, history_start_date: datetime.date = None):
    """
    建立 DB table，並可選擇是否執行初次歷史匯入
    create_history: bool, 若 True 則執行 initial_history_all
    history_start_date: datetime.date 或 None（若 None 預設 2015-01-01）
    """
    engine = get_engine()
    # 建表（get_* 會 create_all）
    get_main_table(engine)
    get_intraday_archive_table(engine)
    print("Database tables ensured.")

    if create_history:
        if history_start_date is None:
            history_start_date = datetime.date(2015, 1, 1)
        print("Start initial history ingestion from", history_start_date)
        initial_history_all(history_start_date)
    else:
        print("Skipping initial history ingestion.")


if __name__ == "__main__":
    # 範例：第一次啟動要抓歷史就把 create_history=True
    setup(create_history=False)
```
