```python
from datetime import datetime, time, timedelta
from zoneinfo import ZoneInfo
from typing import Tuple, Optional

# 使用 America/New_York，會自動處理 DST
NY = ZoneInfo("America/New_York")


def now_ny() -> datetime:
    """回傳現在 NY 時間（timezone-aware，會自動處理 DST）"""
    return datetime.now(NY)


def delayed_timestamp(minutes: int = 15, now: Optional[datetime] = None) -> datetime:
    """回傳延遲 minutes 分鐘的時間點（用於 15min delay）"""
    if now is None:
        now = now_ny()
    if now.tzinfo is None:
        now = now.replace(tzinfo=NY)
    return now - timedelta(minutes=minutes)


def in_work_window(
    start_hm: Tuple[int, int] = (9, 45),
    end_hm: Tuple[int, int] = (16, 15),
    now: Optional[datetime] = None,
) -> bool:
    """
    判斷現在是否在指定工作時段（NY 時區）
    start_hm, end_hm 為 tuple (hour, minute)
    若 now 提供，會以該 datetime（應為 timezone-aware 或 naive）判斷（若 naive 視為 NY 時區）
    """
    if now is None:
        now = now_ny()
    # 若傳入 naive datetime，視為 NY 時區
    if now.tzinfo is None:
        now = now.replace(tzinfo=NY)
    start = time(start_hm[0], start_hm[1])
    end = time(end_hm[0], end_hm[1])
    t = now.timetz()
    return start <= t <= end
```
