```python
import time
import yfinance as yf

symbols = ["AAPL", "MSFT", "NVDA", "AMZN", "GOOGL"]
interval = "15m"
period = "60d"

times = []
for s in symbols:
    t0 = time.time()
    df = yf.download(s, interval=interval, period=period, progress=False)
    t1 = time.time()
    times.append(t1 - t0)
    print(f"{s} took {t1 - t0:.2f}s rows={0 if df is None else len(df)}")

avg = sum(times) / len(times)
print(f"Average time per request: {avg:.2f}s")
```

```python
# if stockanalysis website changed , list all tables and find the right one with index and change to tables[n]
import io
import pandas as pd
import requests

url = "https://stockanalysis.com/list/sp-500-stocks/"
resp = requests.get(url, headers={"User-Agent": "Mozilla/5.0"}, timeout=15)
resp.raise_for_status()

tables = pd.read_html(io.StringIO(resp.text))

# 1) 先列出每個 table 的欄位，快速看哪個像你要的
for i, df in enumerate(tables):
    print(i, list(df.columns))

# 2) 用「條件」找出 table index（不用人工看）
candidates = [i for i, df in enumerate(tables) if "Symbol" in df.columns]
print("tables with Symbol:", candidates)

# 3) 更精準：同時要 Symbol + Market Cap
candidates2 = [i for i, df in enumerate(tables) if {"Symbol", "Market Cap"}.issubset(df.columns)]
print("tables with Symbol & Market Cap:", candidates2)

# 4) 真的看內容（看前幾列）
i = candidates2[0]
display(tables[i].head())
```
