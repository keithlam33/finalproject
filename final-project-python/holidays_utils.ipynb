```python
# holidays_utils.py
import requests
from typing import Dict, Any

# 直接把 Finnhub key 寫在這裡（你要求簡單直接）
FINNHUB_TOKEN = "d65fkg9r01qlmj2v19r0d65fkg9r01qlmj2v19rg"

def market_status() -> Dict[str, Any]:
    """
    呼叫 Finnhub /market/status
    回傳 JSON dict（caller 要依實際回傳欄位檢查）
    """
    url = "https://finnhub.io/api/v1/market/status"
    resp = requests.get(url, params={"token": FINNHUB_TOKEN}, timeout=10)
    resp.raise_for_status()
    return resp.json()

def market_holidays(year: int) -> Dict[str, Any]:
    """呼叫 Finnhub /market-holiday?year=YYYY"""
    url = "https://finnhub.io/api/v1/market-holiday"
    resp = requests.get(url, params={"token": FINNHUB_TOKEN, "year": year}, timeout=10)
    resp.raise_for_status()
    return resp.json()
```
